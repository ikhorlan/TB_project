<h4>TB prevalence by WHO region<h4>

```python
# A pivot table by WHO region
tb_prev_who=tb_prev.pivot_table(index='WHO region', columns='Year', values='Numeric')
tb_prev_who
```

The prevalence rate decreased the most in the group of upper-middle income countries and the least in the high income countries.

```python
tb_prev_who['tb_prev_change']=(tb_prev_who[2014]-tb_prev_who[2007])/tb_prev_who[2007]
tb_prev_who.sort_values('indicator_change', ascending=False)
```

The prevalence of TB decreased the most in the European region and even increased in the Eastern Mediterranean region.

```python
gdp=pd.read_csv('../final_proposal_data/GDP_pcap_2007_2014.csv')
```

```python
gdp.columns=[('year_' if len(i)==4 else '') + i for i in gdp.columns]
gdp.columns
```

```python
gdp.head()
```

```python
gdp=pd.wide_to_long(gdp, stubnames='year_', j='Year', i=['Country']).reset_index().rename(columns={'year_':'GDP_pcap'})
```

```python
gdp.head()
```

```python
tb_prev_gdp=tb_prev.merge(gdp, on=['Country', 'Year'], how='left')
```

```python
tb_prev_gdp.head()
```

```python
le=pd.read_csv('../final_proposal_data/LE2007_2014.csv')
```

```python
le.head()
```

```python
le.columns=[('year_' if len(i)==4 else '') + i for i in le.columns]
le.columns
```

```python
le=pd.wide_to_long(le, stubnames='year_', j='Year', i=['Country']).reset_index().rename(columns={'year_':'Life_exp'})
```

```python
le.head()
```

```python
tb_prev_gdp_le=tb_prev_gdp.merge(le, on=['Country', 'Year'], how='left')
```

```python
tb_prev_gdp_le.dropna(inplace=True)
```

```python
tb_prev_gdp_le.sort_values(by='GDP_pcap', ascending=False).head()
```

```python
hiv=pd.read_csv('../final_proposal_data/HIV_2007_2014.csv')
```

```python
hiv.columns=[('year_' if len(i)==4 else '') + i for i in hiv.columns]
hiv.columns
```

```python
hiv=pd.wide_to_long(hiv, stubnames='year_', j='Year', i=['Country']).reset_index().rename(columns={'year_':'HIV'})
```

```python
hiv.head()
```

```python
tb_prev_gdp_le_hiv=tb_prev_gdp_le.merge(
    hiv, on=['Country', 'Year'], how='left')[[
    'Country', 
    'Year', 
    'WHO region', 
    'World Bank income group', 
    'Numeric', 
    'GDP_pcap', 
    'Life_exp', 'HIV']]
```

```python
tb_prev_gdp_le_hiv.head()
```

```python
tb_prev_gdp_le_hiv.shape
```

```python
tb_prev_gdp_le_hiv['log_gdp_pcap']=np.log10(tb_prev_gdp_le_hiv.GDP_pcap)
```

```python
tb_prev_gdp_le_hiv.sort_values(by='GDP_pcap').tail()
```

```python
dpi=96

for i in tb_prev_gdp_le_hiv.Year.unique():
    plt.ioff()
    fig = plt.figure(figsize=(680/dpi, 480/dpi), dpi=dpi)
    subsetData = tb_prev_gdp_le_hiv[ tb_prev_gdp_le_hiv.Year == i ]

    plt.scatter(
        x=subsetData['log_gdp_pcap'], 
        y=subsetData['Life_exp'], 
        s=subsetData['Numeric'], 
        c=subsetData['World Bank income group'].cat.codes, 
        cmap="Accent", alpha=0.6, edgecolors="white", linewidth=2)
    
    plt.xlabel("GDP per Capita")
    plt.ylabel("Life Expectancy")
    plt.title("Year: "+str(i) )
    plt.ylim(40,100)
    plt.xlim(2, 5)
    
    filename='C:/Users/ikhor/OneDrive/Documents/GA/final_proposal_data/GIF/'+str(i)+'.png'
    plt.savefig(fname=filename, dpi=100, facecolor='white')
    plt.gca()
    plt.close(fig)
```
